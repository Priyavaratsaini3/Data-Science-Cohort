Camparing String

```python
string1 = "apple"
string2 = "banana"

print(string1 == string2) # Equal to
print(string1 != string2) # Not equal to
print(string1 < string2) # Less than
print(string1 > string2) # Greater than
print(string1 <= string2) # less than and equal to
print(string1 >= string2) # greater than and equal to
```

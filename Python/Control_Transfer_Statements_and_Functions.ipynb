```python
i = 1
while i < 3:
  j = 0
  while j < 5:
    j = j + 1
    if j == 3:
      continue
    print(j, end = " ")
  i = i + 1
```

Create a function "calc" in which it is having add functionality, sun func , multi func, and division func, of two numbers

```python
def calc(num1, num2 , operation = 'add'):
  if operation == 'add':
    return num1 + num2
  elif operation == 'sub':
    return num1 - num2
  elif operation == 'mul':
    return num1 * num2
  elif operation == 'div':
    return num1 / num2


result = calc(10, 20, 'mul')
result
```

```python
def division(a,b):
  if b != 0:
    return a / b
  else:
    return "Division by zero is not allowed"

result = division(10, 8)
print(result)
```

```python
def add_with_print(a, b):
  print("The sum of two numbers is", a + b)

def add_with_return(a, b):
  return a + b

add_with_print(2, 3)
result = add_with_return(2, 3)
print("The sum of two numbers is", result)
```

```python
def is_armstrong_number(num):
  num_str = str(num)
  num_digits = len(num_str)
  sum_of_digits = sum(int(digit) ** num_digits for digit in num_str)
  if num == sum_of_digits:
    return True
  else:
    return False

## function calling
result = is_armstrong_number(134)
print(result)
```

```python
## function defination
def sum_and_product(num1, num2):
  sum_result = num1 + num2
  product_result = num1 * num2
  return sum_result, product_result

num1 = 6
num2 = 10
sum_result, product_result = sum_and_product(num1, num2)
print(sum_result)
print(product_result)
```

Nested Function

```python
def calculator(x,y):
  def add(a, b):
    return a + b

  def multiply(a, b):
    return a * b
  def calc(op , a , b):
    if op == 'add':
      return add(a, b)

    else:
      return multiply(a, b)
  return calc('add', x, y), calc('mul', x, y)

result = calculator(2, 993)
print(result)
```

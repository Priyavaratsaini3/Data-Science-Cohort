Arithmetic Operators


```python
x = 10
y = 3

# Preform the arithmetic Operations
print(x + y) # Addition
print(x - y) # Substraction
print(x * y) # Multiplication
print(x / y) # Division
print(x // y) # Floor Division
print(x % y) # Modulus (remainder)
print(x ** y) # Exponentiation
```

```python
a = 10 + 30.34
print(a)
```

Comparison Operators

```python
x = 5
y = 10

# Perform comparison operator
print(x == y) # Equal to
print(x != y) # Not  equal to
print(x > y) # Greater than
print(x < y) # Less than
print(x <= y) # Greater than or equal to
print(x >= y) # Less than or equal to
```

Logical Operators

```python
x = 5
y = 10
z = 15

# Perform logical operators
print(x < y and y < z)
print(x < y or y > z)
print(not(x < y))
```

Bitwise Operator

```python
x = 10
y = 6

# Perform Bitwise Opreations
print(x & y) # AND Oprations
print(x | y) # OR Oprations
print(x ^ y) # XOR Oprations

# task
print(~x) # NOT Opration


print(x << 1) # Left Shift Opration
print(x >> 1) # Right Shift Opration
```

Assingment Operators


```python
x = 20
y = 3

# Perform assingment operators
x += y # x = x + y
print(x)

x -= y # x = x - y
print(x)

x *= y # x = x * y
print(x)

x /= y # x = x / y
print(x)

x %= y # x = x % y
print(x)
```

Membership Operators

```python
print('a' in 'prince')
print('a' not in 'prince')
```

Identity Operators

```python
x = 10
y = 10

# perform identity operation
print(x is y)
print(x is not y)
```

```python
print(id(x))
print(id(y))
```

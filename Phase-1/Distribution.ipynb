```python
import numpy as np
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.stats import skew
```

Data creation for the income

```python
np.random.seed(42)

## Right skewed distribution 
income_distribution = np.random.gamma(shape=2.0, scale=1000.0, size=1000)

## Calculate the skewness 
skew_value = skew(income_distribution)

sns.histplot(income_distribution, bins=30, kde=True)
plt.title(f'Income Distribution (Skewness : {skew_value:.2f})')
plt.xlabel('Income')
plt.ylabel('Frequency')
```

```python
data = pd.read_csv('process_engineering_data_lyst1750493214144.csv')
data.head()
```

Calculate the range of the temperature column in the above data

```python
print(f'Range is {data['temperature'].max() - data['temperature'].min()}')
```

Calculate the IQR of the temperature column in the above data

```python
Q1 = data['temperature'].quantile(0.25)
Q3 = data['temperature'].quantile(0.75)

temperature_IQR = Q3 - Q1

print(temperature_IQR)
```

Calculate the variance of the pressure column in the above data

```python
pressure_variance = data['pressure'].var()
print(pressure_variance)
```

Calculate the standard deviation of the pressure column in the above data.

```python
pressure_std = data['pressure'].std()
print(pressure_std)
```

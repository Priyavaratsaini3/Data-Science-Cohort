Data Creation of the nominal column named 'color'

```python
import pandas as pd
import numpy as np 

np.random.seed(0)
num_samples = 1000
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Crimson', 'Purple']
nominal_data = np.random.choice(colors, num_samples)
nominal_data
```

```python
dataframe = pd.DataFrame(nominal_data, columns=['colors'])
dataframe
```

Data Creation of the ordinal columns named 'satisfaction_levels'

satisfaction_levels = ['low', 'medium', 'high']

```python
satisfaction_levels = ['low', 'medium', 'high']
ordinal_data  = np.random.choice(satisfaction_levels, size=num_samples)
dataframe['satisfaction_levels'] = ordinal_data
dataframe.head()
```

Data creation of the interval data under temperature column

```python
np.random.seed(0)
interval_data = np.random.randint(0, 100, num_samples)
dataframe['tempreature'] = interval_data
dataframe
```

Data creation of the range data under the income column

```python
np.random.seed(0)
income_data = np.random.randint(10000, 1000000, num_samples)
dataframe['income'] = income_data
dataframe
```

Save the file in the csv

```python
dataframe.to_csv('types_of_data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a nominal dataset
data = {
    'Person': ['A', 'B', 'C', 'D', 'E'],
    'Favorite_Color': ['Red', 'Blue', 'Red', 'Green', 'Blue']
}

df = pd.DataFrame(data)

# Step 2: Count frequency of each color
color_counts = df['Favorite_Color'].value_counts()

# # Step 3: Plot the bar chart
plt.figure(figsize=(8, 5),)
color_counts.plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Favorite Colors of People')
plt.xlabel('Color')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample ordinal data (survey ratings)
data = {
    'Customer': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Satisfaction': ['Poor', 'Good', 'Good', 'Excellent', 'Good', 'Average', 'Poor', 'Poor']
}

# Create DataFrame
df = pd.DataFrame(data)

# Define the ordinal order
ordinal_order = ['Poor', 'Average', 'Good', 'Excellent']

# Convert Satisfaction to categorical with order
df['Satisfaction'] = pd.Categorical(df['Satisfaction'], categories=ordinal_order, ordered=True)

# Count of each category
rating_counts = df['Satisfaction'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(8, 5))
sns.barplot(x=rating_counts.index, y=rating_counts.values)

plt.title("Customer Satisfaction Ratings")
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Customers")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample interval data: Average monthly temperatures in Celsius
temperature_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Avg_Temperature_C': [5, 7, 12, 17, 22, 27, 30, 29, 24, 18, 11, 6]
}

# Create DataFrame
df = pd.DataFrame(temperature_data)

# Plot line chart
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Month', y='Avg_Temperature_C', marker='o', color='orange')

plt.title("Average Monthly Temperature (°C)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ratio Data: Weights of 5 people in kg
names = ['Alice', 'Bob', 'Charlie', 'Daisy', 'Ethan']
weights = [55, 72, 68, 80, 60]  # True ratio scale, can be 0

# Stats
average_weight = np.mean(weights)
max_weight = np.max(weights)
min_weight = np.min(weights)

print("Average Weight:", average_weight)
print("Max Weight:", max_weight)
print("Min Weight:", min_weight)

# Visualization
plt.bar(names, weights, color='skyblue')
plt.axhline(y=average_weight, color='red', linestyle='--', label='Average')
plt.title('Weight of Individuals (Ratio Data)')
plt.xlabel('Person')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Data Creation of the nominal column named 'color'

```python
import pandas as pd
import numpy as np 

np.random.seed(0)
num_samples = 1000
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Crimson', 'Purple']
nominal_data = np.random.choice(colors, num_samples)
nominal_data
```

```python
dataframe = pd.DataFrame(nominal_data, columns=['colors'])
dataframe
```

Data Creation of the ordinal columns named 'satisfaction_levels'

satisfaction_levels = ['low', 'medium', 'high']

```python
satisfaction_levels = ['low', 'medium', 'high']
orminal_data  = np.random.choice(satisfaction_levels, size=num_samples)
dataframe['satisfaction_levels'] = orminal_data
dataframe.head()
```

Data creation of the interval data under temperature column

```python
np.random.seed(0)
interval_data = np.random.randint(0, 100, num_samples)
dataframe['tempreature'] = interval_data
dataframe
```

Data creation of the range data under the income column

```python
np.random.seed(0)
income_data = np.random.randint(10000, 1000000, num_samples)
dataframe['income'] = income_data
dataframe
```

Save the file in the csv

```python
dataframe.to_csv('types_of_data.csv')
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

## plot Ratio Data 
plt.hist(dataframe['income'], bins=20, color='green', alpha=0.7, edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income (Rupees)')
plt.ylabel('Frequency')
plt.savefig('income_distribution.png')
plt.show()
```
